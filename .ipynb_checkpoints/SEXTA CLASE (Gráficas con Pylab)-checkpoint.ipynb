```python
"importar libreria Pylab"
import pylab
```

```python
c=16
a=[y for y in range(0,c+1)]
b=[x**2 for x in range(c+1)]
```

```python
a
```

```python
b
```

```python
pylab.plot(a,b)
pylab.xlabel("Eje x")
pylab.ylabel("Eje y")
pylab.title("Funcion Cuadratica")
pylab.show()
```
